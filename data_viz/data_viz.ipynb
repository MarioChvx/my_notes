```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('superstore.xls').sort_values(['Order Date'])
df.head()
```

```python
df_g = df.groupby(['Category'])['Sales'].sum().reset_index()
df_g.head()
```

```python
sns.barplot(data = df_g, y='Sales', x='Category')
```

```python
df_gc = df.groupby(['Customer Name'])['Sales'].sum().reset_index()
df_gc = df.sort_values(['Sales'], ascending = False)
df_gc.head()
```

```python
sns.barplot(data=df_gc.head(10), y='Customer Name', x='Sales')
```

```python
df_m = df.groupby([pd.Grouper(key='Order Date', freq='3M')])['Sales'].sum().reset_index()
```

```python
sns.barplot(data=df_m, y='Order Date', x='Sales')
```
