# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

## subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
axes[0][0].plot(x,y,'--')
axes[0][1].plot(y,x,'r--')
fig.tight_layout()
```

```python
fig, axes = plt.subplots(1,2, figsize = (10, 5))
axes[0].plot(x,y,label="$sin(x)$")
axes[0].set_title('Relacion x - y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[0].plot(x,y,label="sin(x)")
axes[0].set_title('Relacion y - x')
axes[1].set_xlabel('Velocidad')
axes[1].set_ylabel('Tiempo')
axes[0].legend()
```

## Bar plot 

```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.bar(countrys, population)
plt.show()
```

```python
plt.bar(countrys, population, width=.5, color=['aqua', 'grey', 'teal', 'crimson', 'violet'])
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemanaia'), rotation = 35)
plt.show()
```

```python
plt.barh(countrys, population,  color=['aqua', 'grey', 'teal', 'crimson', 'violet'])
plt.yticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemanaia'), rotation = 35)
plt.show()
```
