# scraping

```python
import requests

response = requests.get('https://www.milenio.com')
print(response.status_code)
```

```python
print(response.headers)
```

```python
for key, value in response.headers.items():
    print(f"{key} : {value}")
```

```python
print(response.text)
```

## beautifulsoup

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(response.text, 'html.parser')

print(soup.title.text)
```

```python
print(soup.select('meta[name=description]'))
```

```python
news_links = soup.select('.sn-base-base')

for new in news_links:
    print(new.contents)
```
