# scraping

```python
import requests
milenio = 'https://www.milenio.com'
response = requests.get(milenio)
print(response.status_code)
```

```python
print(response.headers)
```

```python
for key, value in response.headers.items():
    print(f"{key} : {value}")
```

```python
print(response.text)
```

## beautifulsoup

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(response.text, 'html.parser')

print(soup.title.text)
```

```python
print(soup.select('meta[name=description]'))
```

```python
news_links = []
news_links = soup.select('.sn-base-base-noheading a')
news_links = [new['href'] for new in news_links]
news_links = [milenio + link for link in news_links]
news_content = [requests.get(link) for link in news_links]
```

```python
news_content = [bs4.BeautifulSoup(page.text, 'html.parser') for page in news_content]
```

```python
news_content = [{'title' : page.select('.nd-title-headline-title-headline-base__title'), 'content' : page.select('.media-container.news')} for page in news_content]
```

```python
news_content = [{'title' : new['title'][0].string, 'content' : new['content'][0].find_all('p')} for new in news_content]
```

```python
for p in news_content[0]['content']:
    for string in p.stripped_strings:
        print(string)
    print('~~~~~~~~~~~~')
```

```python
test = ''.join([string for string in news_content[0]['content'][0].stripped_strings])
print(test)
```
