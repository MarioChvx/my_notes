# seaborn

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## set

```python
sns.barplot(x=['A','B','C'], y=[1,2,3])
plt.show()
```

```python
sns.set(style='darkgrid', palette='dark', font='Verdana', font_scale=2)
sns.barplot(x=['A','B','C'], y=[1,2,3])
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='time')
plt.show()
```

## distribuciones

```python
tips.head()
```

```python
sns.histplot(data=tips, x='tip', bins=10, cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde')
plt.show()
```

## cualitativos

```python
tips.head()
```

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.show()
```

## relation

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time',size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time',size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time',size='size', col='time')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

## joinplot pairplot

```python
sns.jointplot(data=tips, x='total_bill', y='tip',hue='sex',kind='hist', marginal_ticks=True,
            marginal_kws=dict(bins=25, multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tips, hue='smoker', kind='reg')
plt.show()
```
