## Merge, join and concat

### concat

```python
df1 = pd.DataFrame({
	'A':['A0', 'A1', 'A2','A3'],
    'B':['B0', 'B1', 'B2','B3'],
	'C':['C0', 'C1', 'C2','C3'],
	'D':['D0', 'D1', 'D2','D3']
})

df2 = pd.DataFrame({
	'A':['A4', 'A5', 'A6','A7'],
	'B':['B4', 'B5', 'B6','B7'],
	'C':['C4', 'C5', 'C6','C7'],
	'D':['D4', 'D5', 'D6','D7']
})
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1, ignore_index=True)
```

### Merge and Join

join just work with index values

```python
izq = pd.DataFrame({
    'key':['k0', 'k1', 'k2','k3'],
    'A' : ['A0', 'A1', 'A2','A3'],
    'B' : ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({
    'key':['k0', 'k1', 'k2','k3'],
    'C' : ['C0', 'C1', 'C2','C3'],
    'D' : ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({
    'key':['k0', 'k1', 'k2','k3'],
    'A' : ['A0', 'A1', 'A2','A3'],
    'B' : ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({
    'kie':['k0', 'k1', 'k2','k3'],
    'C' : ['C0', 'C1', 'C2','C3'],
    'D' : ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, left_on= 'key', right_on='kie')
```

```python
izq = pd.DataFrame({
    'key':['k0', 'k1', 'k2','k3'],
    'A' : ['A0', 'A1', 'A2','A3'],
    'B' : ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({
    'kie':['k2', 'k3', 'k4','k5'],
    'C' : ['C0', 'C1', 'C2','C3'],
    'D' : ['D0', 'D1', 'D2','D3']})
```

```python
izq.merge(der, left_on= 'key', right_on='kie', how='left')
```

```python
izq.merge(der, left_on= 'key', right_on='kie', how='right')
```
